```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()  # Load environment variables from .env file
```

```python
model = ChatOpenAI()
```

```python
# define the state structure
class LLMState(TypedDict):
    prompt: str
    response: str
```

```python
def llm_workflow(state: LLMState) -> LLMState:

    # extract prompt from state
    prompt = state["prompt"]

    # form a prompt for LLM
    llm_prompt = f"Please provide a detailed response to the following prompt:\n\n{prompt}"

    # get response from LLM
    response = model.invoke(llm_prompt).content

    # update state with response
    state["response"] = response

    return state
```

```python
# create the graph
graph = StateGraph(LLMState)

# add node
graph.add_node("llm_workflow", llm_workflow)

# add edges to graph
graph.add_edge(START, "llm_workflow")
graph.add_edge("llm_workflow", END)

# compile the graph
workflow = graph.compile()
```

```python
# execute the workflow
Initial_State = {'prompt': 'Explain the theory of relativity.'}
final_state = workflow.invoke(Initial_State)
print(final_state)  # Output will include the LLM response
```
